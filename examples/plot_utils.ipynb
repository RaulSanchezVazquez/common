Add freq. values on top of bar-plots

```python
%matplotlib inline

import pandas as pd
from sklearn import datasets

from common.plot_utils import add_bar_values

iris = datasets.load_iris()
y = pd.Series([iris.target_names[x] for x in iris.target])
y_vc = y.value_counts()

y_vc
```

```python
ax = y_vc.plot(kind='bar', figsize=(6, 3), ylim=(0, 60), grid=True)
add_bar_values(ax, kind='bar');
```

```python
ax = y_vc.plot(kind='barh', figsize=(6, 3), ylim=(0, 60), grid=True)
add_bar_values(ax, kind='barh');
```
