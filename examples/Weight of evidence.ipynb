```python
import matplotlib.pyplot as plt
import pandas as pd

y = pd.Series(
    (['Event'] * 150) + (['No Event'] * 1050) + \
    (['Event'] * 120) + (['No Event'] * 780) + \
    (['Event'] * 110) + (['No Event'] * 980) + \
    (['Event'] * 100) + (['No Event'] * 1360) + \
    (['Event'] * 50) + (['No Event'] * 1360)
)

X = pd.DataFrame({
    'age': (
        [20] * 1200
    ) + (
        [30] * 900
    ) + (
        [40] * 1090
    ) + (
        [50] * 1460
    ) + (
        [60] * 1410
    )
})
```

```python
X['age'].value_counts().sort_index().plot(kind='bar', grid=True);
```

```python
y.value_counts()
```

Set to `str` whatever you wish to become a feature encoded as WOE.

```python
X['age'] = X['age'].astype(str)
```

Also use the binary target encoded as `{0, 1}`

```python
y = (y == 'No Event').astype(int)
```

```python
from common import weight_of_evidence

woe_scaler = weight_of_evidence.WOES()
woe_scaler.fit(X, y)
```

```python
woe_scaler.woes
```

```python
X = woe_scaler.transform(X)
X
```
